Imagina que te llega un conjunto de datos que contiene tres variables categóricas>> A,B y C (con 2 categorías cada una). ¿Podrías plantear una estrategia que recoja los 3 factores realizando un Two-Way ANOVA? ¿y un One-Way ANOVA? ¿Qué consideraciones tendrías en cuenta al diseñar estos análisis? (2pt)

El **Two-Way ANOVA** s'utilitza per analitzar la influència de dues variables categòriques (anomenades factors) en una variable contínua. En aquest cas, hi han tres variables categòriques (A, B i C).

**Estratègia per realitzar un Two-Way ANOVA:**

**1. Hipòtesi nul·la i alternativa:**
- H0 (Hipòtesi nul·la): No hi ha efectes significatius de les variables A, B o la interacció AB i BC.
- H1 (Hipòtesi alternativa): Hi ha almenys un efecte significatiu de les variables A, B o la interacció AB i BC.

**2. Preparació de dades:**
- Assegura't que les dades estiguin ben organitzades, amb les observacions classificades segons les categories de A, B i C.

**3. Anàlisi de la variància (ANOVA):**

- Realitza un Two-Way ANOVA.
- Divideix la variabilitat total en tres components: efecte de A, efecte de B i la interacció AB.
- Examina les estadístiques F i els seus p-values associats per a cada factor i la interacció.

**4. Proves post-hoc (si és necessari):**
- Si hi han efectes significatius, es poden realitzar proves post-hoc per determinar quins nivells dels factors són diferents.

**5. Interpretació de resultats:**

- Analitza els resultats en funció de les estadístiques F i els p-values.
- Si hi ha efectes significatius, interpreta la direcció d'aquests efectes i la seva rellevància pràctica.

**Consideracions:**
- Assegura que les dades compleixin amb les suposicions de normalitat i homogeneïtat de variàncies.
- Es pot realitzar transformacions de dades si és necessari per complir amb les suposicions.


El **One-Way ANOVA** s'utilitza per comparar les mitjanes de tres o més grups. En aquest cas, cada variable categòrica (A, B, C) es pot considerar com un grup.

**Estratègia per realitzar un One-Way ANOVA:**

**1. Hipòtesi nul·la i alternativa:**
- H0: No hi ha diferències significatives entre les mitjanes dels grups A, B i C.
- H1: Hi ha almenys una diferència significativa entre les mitjanes dels grups A, B i C.

**2. Preparació de dades:**
- Organitza les dades de manera que es puguin comparar les mitjanes de les variables A, B i C.

**3. Anàlisi de la variància (ANOVA):**
- Realitza un One-Way ANOVA
- Calcula les sumes de quadrats i els graus de llibertat pel factor (A, B, C).
- Examina l'estadística F i el seu p-value associat.

**4. Proves post-hoc (si és necessari):**
- Si l'ANOVA mostra diferències significatives, es poden realitzar proves post-hoc per identificar quins grups són diferents entre ells.

**5. Interpretació de resultats:**

- Analitza els resultats en funció de les estadístiques F i els p-values.
- Si hi ha efectes significatius, interpreta la direcció d'aquests efectes i la seva rellevància pràctica.

**Consideracions:**
- Verifica les suposicions de normalitat i homogeneïtat de variàncies.
- Aplicar transformacions si és necessari.


En el disseny d'aquests anàlisis, és important tenir en compte diverses consideracions. En primer lloc, s'han de verificar les suposicions de normalitat i homogeneïtat de variàncies, utilitzant proves estadístiques específiques com Shapiro-Wilk (normalitat) i d'homogeneïtat de variàncies.

Transformacions a les dades si no es compleixen les suposicions, com ara el logaritme, que es transformacio comuna. En cas de trobar efectes significatius, es poden realitza proves post-hoc per obtenir informació detallada sobre les diferències entre els grups.

# 1. Visualización y comprensión del dataset (2p):

Se pide visualizar y realizar estadísticos que ayuden a comprender mejor el dataset dentro del dominio de ANOVA: Análisis descriptivo de las variables continuas y de nuestros factores y análisis de correlaciones


```r
# Visualización y estadísticos descriptivos de variables continuas
summary(data_ais[, c("rcc", "wcc", "hc", "hg", "bmi")])
```

Cálculo de estadísticas descriptivas

```r

stats_numericas <- data_ais %>%
  summarize(
    m_rcc = mean(rcc), sd_rcc = sd(rcc),
    m_wcc = mean(wcc), sd_wcc = sd(wcc),
    m_hc = mean(hc), sd_hc = sd(hc),
    m_hg = mean(hg), sd_hg = sd(hg),
    m_bmi = mean(bmi), sd_bmi = sd(bmi)
  )
```

Resumen por género y deporte

```r
summary_by_gender_sport <- data_ais %>%
  group_by(sex, sport) %>%
  summarize(
    m_rcc = mean(rcc), sd_rcc = sd(rcc),
    m_wcc = mean(wcc), sd_wcc = sd(wcc),
    m_hc = mean(hc), sd_hc = sd(hc),
    m_hg = mean(hg), sd_hg = sd(hg),
    m_bmi = mean(bmi), sd_bmi = sd(bmi)
  )
```

Diferencias entre géneros y deportes

```r
diff_row <- summary_by_gender_sport %>%
  summarise(sex = "diff_sex_sport",
            m_rcc = diff(m_rcc), sd_rcc = diff(sd_rcc),
            m_wcc = diff(m_wcc), sd_wcc = diff(sd_wcc),
            m_hc = diff(m_hc), sd_hc = diff(sd_hc),
            m_hg = diff(m_hg), sd_hg = diff(sd_hg),
            m_bmi = diff(m_bmi), sd_bmi = diff(sd_bmi)
  )
```

Histogramas

```r
# Combinar resultados
summary_results <- bind_rows(summary_by_gender_sport, diff_row)

# Visualización - Histogramas
graficos_histograma <- list()
var_data <- colnames(data_ais[, c("rcc", "wcc", "hc", "hg", "bmi")])

for (variable in var_data) {
  grafico <- ggplot(data_ais) +
    aes_string(x = variable) +
    geom_histogram(bins = 8, fill = '#e69138', color = 'white', na.rm = TRUE) +
    theme(text = element_text(size = 15), axis.text.x = element_text(size = 14)) +
    labs(title = paste('Hist', variable))

  graficos_histograma[[length(graficos_histograma) + 1]] <- grafico
}

options(
  repr.plot.width = 15,
  repr.plot.height = 10
)

# Mostrar los histogramas en un grid
do.call('grid.arrange', c(graficos_histograma, ncol = 3))
```

```r
# Boxplots
box_plots <- list()
var_data <- colnames(data_ais[, c("rcc", "wcc", "hc", "hg", "bmi")])

for (variable in var_data) {
  box_plot <- ggplot(data_ais, aes(x = '', y = !!sym(variable))) +
    stat_boxplot(geom = 'errorbar', color = '#3c0023', na.rm = TRUE) +
    geom_boxplot(fill = '#e69138', color = '#3c0023', na.rm = TRUE) +
    coord_flip() +
    labs(title = paste('Boxplot', variable), y = variable) +
    theme(panel.background = element_rect(fill = 'lightgray'),
          panel.grid.major = element_line(color = 'white'),
          panel.grid.minor = element_line(color = 'white')) +
    theme(text = element_text(size = 12))
  box_plots[[variable]] <- box_plot
}

options(
  repr.plot.width = 15,
  repr.plot.height = 10
)
grid.arrange(grobs = box_plots, ncol = 3, widths = rep(2, 3))

```

```r
# Q-Q Plots
options(
  repr.plot.width = 15,
  repr.plot.height = 8
)
qq_plots <- list()
var_data <- colnames(data_ais[, c("rcc", "wcc", "hc", "hg", "bmi")])

for (i in seq_along(var_data)) {
  variable <- var_data[i]
  qq_plot <- ggplot(data_ais[!is.na(data_ais[[variable]]), ],aes(sample = !!sym(variable))) +
    stat_qq() +
    stat_qq_line(color ='#e69138', lwd = 1) +
    labs(title = paste('qqplot', variable),
         x = 'Theoretical Quantiles',
         y = 'Sample Quantiles') +
    theme_classic() +
    theme(text = element_text(size = 15))
  qq_plots[[i]] <- qq_plot
}

grid.arrange(grobs = qq_plots, ncol = 4, widths = rep(2, 4))
```

# 2. Calcular la prueba ANOVA One-Way para la variable sport respecto 2 variables continuas (escoged).
Interpretar resultados y acompañar con los gráficos vistos en clase (3p)


```r
# ANOVA para la variable 'sport' respecto a 'rcc' y 'wcc'
modelo_anova_sport <- aov(cbind(rcc, wcc) ~ sport, data = data_ais)

# Resumen del modelo ANOVA
summary(modelo_anova_sport)
```
