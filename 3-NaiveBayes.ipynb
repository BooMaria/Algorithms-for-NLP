3.Naive Bayes

És una tècnica de classificació basada en el **teorema** de **Bayes** amb un supòsit d'**independència** entre predictors. 

Un classificador Naive Bayes suposa que la presència d'una **característica** **particular** en una classe no està relacionada amb la presència de cap altra característica.



No és un únic algorisme sinó una **família d'algorismes** on comparteixen un **principi comú**, és a dir, cada parell de característiques que es classifiquen són **independents** les unes de les altres.

El model Naive Bayes és **fàcil de construir** i especialment útil per a **data sets** molt **grans**. 
Al ser **molt simplificat**, Naive Bayes supera fins i tot els mètodes de classificació molt sofisticats.


image.png




**P(c|x)** és la probabilitat posterior de la classe (c, objectiu) donat el predictor (x, atributs).

**P(c)** és la probabilitat prèvia de la classe.

**P(x|c)** és la probabilitat que és la probabilitat de predictor donada classe.

**P(x)** és la probabilitat prèvia del predictor.

**Applicacions: **
-	Predicció en temps real.
-	Predicció multiclasse.
-	Classificació de text/Filtret de correu brossa/Anàlisi de sentiments.
-	Sistema de recomanació.



Hi ha tres tipus de **model** **Naive Bayes **a la biblioteca scikit-learn: 

**Gaussian**: s'utilitza en classificació i suposa que les característiques segueixen una **distribució normal**

**Multinomial**: S'utilitza per a **discrete counts**. (comptar la freqüència amb què apareix la paraula al document).

**Bernoulli**: el model binomial és útil si els vectors de característiques són **binaris**.


**Pros: **
-	És **fàcil** i **ràpid** predir la classe del test **data set**. 
-	També funciona bé en predicció **multiclasse**. 

-	Funciona millor en comparació amb altres models com ara la regressió logística i necessiteu **menys dades d'entrenament**.

-	Funciona bé en cas de **variables d'entrada categòriques** en comparació amb variables numèriques. 


**Cons: **

-	Si la variable categòrica té una **categoria que no s'ha observat** en el conjunt de dades d'entrenament, el model assignarà una probabilitat 0 (zero) i no podrà fer una predicció. Per solucionar-ho, podem utilitzar la tècnica de **suavització**. Una de les tècniques de suavització més senzilles s'anomena **estimació de Laplace.**

-	Una altra limitació de Naive Bayes és **l'assumpció de predictors** independents. 
A la vida real, és gairebé impossible que tinguem un conjunt de predictors que siguin completament independents.

# Codi

Assigning features and label variables

```python

wheather=['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny','Sunny',
'Rainy','Sunny','Overcast','Overcast','Rainy']
temp=['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild']

play=['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']
```

Import LabelEncoder

```python

from sklearn import preprocessing
```

creating labelEncoder

```python
le = preprocessing.LabelEncoder()
```

Converting string labels into numbers.

```python
wheather_encoded = le.fit_transform(wheather)
print( wheather_encoded)
```

Converting string labels into numbers

```python
temp_encoded=le.fit_transform(temp)
label=le.fit_transform(play)
print("Temp:",temp_encoded)
print("Play:",label)
```

Combinig weather and temp into single listof tuples

```python
features=list(zip(wheather_encoded,temp_encoded))

print(features)
```

# Import Gaussian Naive Bayes model

```python
from sklearn.naive_bayes import GaussianNB
```

Create a Gaussian Classifier

```python
model = GaussianNB()
```

Train the model using the training sets

```python
model.fit(features,label)
```

Predict Output

```python
predicted= model.predict([[0,2]]) # 0:Overcast, 2:Mild
print("Predicted Value:", predicted)
```

# Naive Bayes with Multiple Labels

Import scikit-learn dataset library



```python
from sklearn import datasets
```

Load dataset

```python
wine = datasets.load_wine()
```

print the names of the 13 features
print the label type of wine(class_0, class_1, class_2)

```python
print("Features: ", wine.feature_names)
print("Labels: ", wine.target_names)
```

print data(feature)shape


```python
wine.data.shape
```

Print the wine data features (top 5 records)

Print the wine labels (0:Class_0, 1:class_2, 2:class_2)


```python
print(wine.data[0:5])
print(wine.target)
```

Import train_test_split function

```python
from sklearn.model_selection import train_test_split
```

Split dataset into training set and test set

```python
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.3,random_state=109)
# 70% training and 30% test
```

Import Gaussian Naive Bayes model

```python
from sklearn.naive_bayes import GaussianNB
```

Create a Gaussian Classifier

```python
gnb = GaussianNB()
```

Train the model using the training sets

```python
gnb.fit(X_train, y_train)
```

Predict the response for test dataset

```python
y_pred = gnb.predict(X_test)
```

Import scikit-learn metrics module for accuracy calculation

```python
from sklearn import metrics
```

Model Accuracy, how often is the classifier correct?

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
