1. Planteja un disseny de test d'hipòtesis

# **Tema del disseny:** Personalització en la interacció dels chatbots i el seu impacte en la satisfacció de l'usuari en comparació amb respostes predefinides.

**Hipòtesi nul·la (H0):** No hi ha una diferència significativa en la satisfacció de l'usuari entre la interacció personalitzada dels chatbots i les respostes predefinides.

**Hipòtesi alternativa (H1):** La interacció personalitzada dels chatbots resulta en una major satisfacció de l'usuari en comparació amb les respostes predefinides.



# **Disseny del Test d'Hipòtesis:**

**Selecció de Mostra:**
- Selecciona aleatòriament dos grups d'usuaris.
- Grup A: Interactuarà amb un chatbot que proporciona respostes predefinides.
- Grup B: Interactuarà amb un chatbot que adapta les seves respostes segons l'historial i les preferències de l'usuari.

**Variable Dependent:**
- Satisfacció de l'usuari, mesurada mitjançant enquestes després de la interacció amb el chatbot.

**Variable Independent:**
- Tipus d'interacció del chatbot (predefinida vs. personalitzada).

**Procediment:**
- Tots dos grups s'enfronten a situacions similars i realitzen tasques específiques a través del chatbot.
- Després de la interacció, els usuaris completen enquestes de satisfacció que inclouen preguntes sobre la utilitat, l'eficàcia i l'experiència general.

**Anàlisi Estadística:**
- Utilitzar una prova estadística, com ara la prova t de Student, per comparar les puntuacions de satisfacció entre els dos grups.
- Establablir un nivell de significança (p. ex., 0,05) per determinar si la diferència és estadísticament significativa.

**Controls i Variables Confuses:**
- Controlar variables com la complexitat de les consultes, la freqüència d'ús del chatbot i la familiaritat prèvia amb tecnologia similar.
- Assegura't que la mostra sigui representativa de la població objectiu.

**Conclusions:**
- Si el valor p és menor que el nivell de significança, es rebutja la hipòtesi nul·la i accepta la hipòtesi alternativa.
- Si el valor p és major que el nivell de significança, no hi ha prou evidència per rebutjar la hipòtesi nul·la.

```r
# Dividir les dades en dues mostres segons la variable "Group"
group_0 <- data$Order.Value[data$Group == 0]
group_1 <- data$Order.Value[data$Group == 1]

# Test de t-Student independent
t_test_result <- t.test(group_0, group_1)

# Imprimir els resultats
print(t_test_result)

# Visualitzar l'histograma amb la zona de rebuig
hist(group_0, col="lightblue", main="Histograma de Order.Value per Group",
     xlab="Order.Value", ylab="Frequència", xlim=c(0, 10000))
hist(group_1, col="lightgreen", add=TRUE)
legend("topright", legend=c("Group 0", "Group 1"), fill=c("lightblue", "lightgreen"))

# Afegir línies per la zona de rebuig
abline(v = t_test_result$conf.int, col="red", lty=2)
```
